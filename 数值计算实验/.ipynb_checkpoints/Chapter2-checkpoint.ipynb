## 题目1

求方程$2x^2+x-15=0$的正根（$x^*=2.5$）近似值，分别用如下三种格式编程计算：
* $x_{k+1}=15-x_k^2,k=0,1,2,\cdots$取初始值$x_0=2$
* $x_{k+1}=\frac{15}{2x_k+1},k=0,1,2,\cdots$取初始值$x_0=2$
* $x_{k+1}=x_k-\frac{2x_k^2+x_k-15}{4x_k+1},k=0,1,2,\cdots$取初始值$x_0=2$

依次计算$x_1,x_2,\cdots,x_k,\cdots,$并作图观察解的稳定性、收敛性并分析其原因

### 实验代码

库的调用与设置

```python
import numpy as np

from matplotlib import pyplot as plt


plt.rcParams['figure.dpi'] = 80
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

$x_{k+1}=15-x_k^2,k=0,1,2,\cdots$取初始值$x_0=2$

```python
x = [2]
y = [0]
for i in range(1, 5):
    # x.append(15/(2*x[i-1]+1))
    x.append(15-x[i-1]*x[i-1])
    # fm=x[i-1]-((2*x[i-1]*x[i-1])+x[i-1]-15
    # x.append(x[i-1]-(2*x[i-1]*x[i-1]+x[i-1]-15)/(4*x[i-1]+1))
    print(x[i])
    y.append('x'+str(i))

plt.title("迭代公式为：$x_{k+1}=15-x_k^2$时")
#plt.xlabel("x的编号", fontproperties=zh_font)
#plt.ylabel("xk值的变化", fontproperties=zh_font)
plt.plot(y[1:], x[1:])
plt.show()
```

$x_{k+1}=\frac{15}{2x_k+1},k=0,1,2,\cdots$取初始值$x_0=2$



```python
x = [2]
y = [0]
for i in range(1, 50):
    x.append(15/(2*x[i-1]+1))
    #print(x[i])
    y.append('x'+str(i))

plt.title("迭代公式为：$x_{k+1}= 15 /(2x_k+1)$时")
#plt.xlabel("x的编号", fontproperties=zh_font)
#plt.ylabel("xk值的变化", fontproperties=zh_font)
plt.plot(y[1:], x[1:])
plt.show()
```

$x_{k+1}=x_k-\frac{2x_k^2+x_k-15}{4x_k+1},k=0,1,2,\cdots$取初始值$x_0=2$

```python
x = [2]
y = [0]
for i in range(1, 50):
    x.append(x[i-1]-(2*x[i-1]*x[i-1]+x[i-1]-15)/(4*x[i-1]+1))
    #print(x[i])
    y.append('x'+str(i))

plt.title("迭代公式为：$x_{k+1}=x_k-(2x_k^2+x_k-15)/(4x_k+1)$时")
#plt.xlabel("x的编号", fontproperties=zh_font)
#plt.ylabel("xk值的变化", fontproperties=zh_font)
plt.xticks([])
plt.plot(y[1:], x[1:])
plt.show()
```
