```python
# -*- coding=utf-8 -*-
import numpy as np
from matplotlib import pyplot as plt

x = [2]
y = [0]
for i in range(1, 10):
    # x.append(15/(2*x[i-1]+1))
    x.append(15-2*x[i-1]*x[i-1])
    # fm=x[i-1]-((2*x[i-1]*x[i-1])+x[i-1]-15
    # x.append(x[i-1]-(2*x[i-1]*x[i-1]+x[i-1]-15)/(4*x[i-1]+1))
    print(x[i])
    y.append('x'+str(i))

plt.title("Matplotlib demo")
#plt.xlabel("x的编号", fontproperties=zh_font)
#plt.ylabel("xk值的变化", fontproperties=zh_font)
plt.plot(y[1:], x[1:])
plt.show()
```
