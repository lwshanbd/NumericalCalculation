```python
import scipy.interpolate
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline

x = np.linspace(0, np.pi, 11)
y = np.sin(x)
xx = np.linspace(x.min(), x.max(), 50)
#xx = np.linspace(x.min(), x.max(), 11)
fig, ax = plt.subplots(figsize=(8, 4))
ax.scatter(x, y)
for s in ['not-a-knot', 'clamped', 'natural', 'periodic']:
    f = CubicSpline(x, y, bc_type=s)
    plt.scatter(xx, f(xx), label=s, lw=0.5,s=3)
    plt.fill_between(xx,f(xx) , alpha = 0.1)
    
plt.scatter(xx,np.sin(xx),label='$sin(x)$',lw=0.3,s=3)
print(f(0.5))
plt.legend()
plt.savefig('4-1.pdf', dpi=250)
plt.rcParams['figure.dpi'] = 250
plt.show()





```
